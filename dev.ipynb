```python
import matplotlib.pyplot as plt
import cv2
```

# Load Image

```python
img_path = "./examples/meme.jpg"
```

```python
image = cv2.imread(img_path) # cv2 mostly read image as in BGR color format
```

```python
type(image)
```

```python
# image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

def plot_img(img, title="meme", gray=False):
    if gray==True:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        plt.imshow(img, cmap="gray")
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # matplotlib and PIL using RGB color format
        plt.imshow(img)        
    plt.axis('off')
    plt.title("Meme")
    plt.show()

plot_img(image)
```

# Image Transformation

## Resize

```python
image.shape
```

```python
n = 244
resized_img = cv2.resize(image, (n,n))
plot_img(resized_img)
```

## Gray

```python
plot_img(image, gray=True)
```

## Horizontal Flip

```python
flip = cv2.flip(image, 1)
plot_img(flip)
```

## Vertical Flip

```python
flip = cv2.flip(image, 0)
plot_img(flip)
```
