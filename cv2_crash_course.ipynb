```python
import matplotlib.pyplot as plt
import cv2
```

# Load Image

```python
img_path = "./examples/meme.jpg"
```

```python
image = cv2.imread(img_path) # cv2 mostly read image as in BGR color format
```

```python
type(image)
```

```python
# image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) 

def plot_img(img, title="meme", gray=False):
    #     img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # matplotlib and PIL using RGB color format
    params = {}
    if gray==True:
        params = {"cmap":"gray"}
    plt.imshow(img, **params)
    plt.axis('off')
    plt.title(title)
    plt.show()


plot_img(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```

# Image Transformation

## Resize

```python
image.shape
```

```python
n = 244
resized_img = cv2.resize(cv2.cvtColor(image, cv2.COLOR_BGR2RGB), (n,n))
plot_img(resized_img)
```

## Gray

```python
image.shape
```

```python
plot_img(cv2.cvtColor(image, cv2.COLOR_BGR2GRAY), gray=True)
```

## Horizontal Flip

```python
flip = cv2.flip(image, 1)
plot_img(cv2.cvtColor(flip, cv2.COLOR_BGR2RGB))
```

## Vertical Flip

```python
flip = cv2.flip(image, 0)
plot_img(cv2.cvtColor(flip, cv2.COLOR_BGR2RGB))
```

## Equalize Histogram

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plot_img(cv2.equalizeHist(gray_img), gray=True)
```

## Brightness and Contrast

### Brightness

```python
# Parameters for adjustment
alpha = 1  # Contrast control (1.0-3.0)
beta = 100    # Brightness control (0-100)

# Apply the adjustments
adjusted_image = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)

plot_img(cv2.cvtColor(adjusted_image, cv2.COLOR_BGR2RGB))
```

# Clipping

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an example image
image = cv2.imread(img_path)

# Over-apply brightness (clipping example)
bright_image_clipped = image + 150  # Adds 100 to every pixel (may exceed 255)

# Over-apply contrast (clipping example)
contrast_image_clipped = image * 5  # Doubles every pixel value (may exceed 255)

# Correct adjustment using cv2.convertScaleAbs
bright_image_corrected = cv2.convertScaleAbs(image, alpha=1, beta=100)
contrast_image_corrected = cv2.convertScaleAbs(image, alpha=2, beta=0)

# Plot the images
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(bright_image_clipped.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title("Over-Brightened (Clipping)")
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(cv2.cvtColor(contrast_image_clipped.astype(np.uint8), cv2.COLOR_BGR2RGB))
plt.title("Over-Contrasted (Clipping)")
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(cv2.cvtColor(bright_image_corrected, cv2.COLOR_BGR2RGB))
plt.title("Corrected Brightness")
plt.axis('off')

plt.subplot(2, 3, 6)
plt.imshow(cv2.cvtColor(contrast_image_corrected, cv2.COLOR_BGR2RGB))
plt.title("Corrected Contrast")
plt.axis('off')
plt.show()
```
